```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('Sonar Data.csv',header=None)
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

```python
X=data.drop(columns=60,axis=1)
Y=data[60]
```

```python
print("Featuers:",X)
print("Output:",Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=4)
```

```python
print(x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
train_prediction=model.predict(x_train)
train_accuracy=accuracy_score(train_prediction,y_train)
```

```python
print("Accuracy of Training Data is:",train_accuracy)
```

```python
test_prediction=model.predict(x_test)
test_accuracy=accuracy_score(test_prediction,y_test)
```

```python
print("Accuracy of Testing Data is:",test_accuracy)
```
